```python
import pandas as pd
import numpy as np
```

# Preparation

- Impoart Raw Data
- Create relavent dictionaries

```python
# Create a dictionary to record manipulated data
EP                                          = {}
df_dict                                     = {}

# Dataframe of raw data from ASPEN
T                                           = [720,740,760,780,800,820]
P                                           = [1,2,3,4,5,6]

for i in T:
    df_dict[f'T={i}K,P=6atm']               = pd.read_excel('Book1.xlsx',sheet_name=f'T={i}K,P=6atm')

for i in P:
    df_dict[f'T=820K,P={i}atm']             = pd.read_excel('Book1.xlsx',sheet_name=f'T=820K,P={i}atm')

for i in df_dict.keys():
    EP[i]                                   = {}    

df                                          = pd.read_excel('Book1.xlsx',sheet_name='T=820K,P=6atm')

# Permanently changes the pandas settings
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Display Dataframe

df_dict['T=820K,P=6atm']
```

# Reactor + Catalyst Cost

```python
# Parameters
MandS                                   = 1800
Fc                                      = 2.3625
bed_voyage                              = 0.45
catalyst_density                        = 1310
price_catalyst                          = 1.1
plant_runtime                           = 8150

for i in df_dict.keys():
    df = df_dict[i]

    # Reactor Installation Cost
    installation_cost                   = (MandS/280)*101.3 * (df['Diameter']*3.281)**1.066* (df['Length']*3.281)**0.802 * (2.18+Fc)
    total_capital_investment            = 2.096*installation_cost
    annual_capital_investment           = 1/3*total_capital_investment

    # Reactor Catalyst Operating Costs
    reactor_volume                      = np.pi*df['Diameter']**2*df['Length']/4
    catalyst_cost                       = reactor_volume*(1-bed_voyage)*catalyst_density*price_catalyst
    annual_catalyst_cost                = catalyst_cost * plant_runtime/96

    # Total annual reactor cost
    annual_reactor_cost                 = annual_capital_investment+annual_catalyst_cost

    # Create a dataframe for reactor cost
    Reactor_Cost = pd.DataFrame({
        'Reactor_Cost'                  : {},
        'XTOL'                          : df['XTOL'],
        'Diameter(ft)'                  : df['Diameter']*3.281,
        'Length(ft)'                    : df['Length']*3.281,
        'installation_cost'             : installation_cost,
        "total_capital_investment"      : total_capital_investment,
        "annual_capital_investment"     : annual_capital_investment,
        "catalyst_cost"                 : catalyst_cost,
        "annual_catalyst_cost"          : annual_catalyst_cost,
        'annual_reactor_cost'           : annual_reactor_cost
    })

    # Save the dataframe in the dictionary
    EP[i]['Reactor_Cost']                  = Reactor_Cost

# Display 
EP['T=820K,P=6atm']['Reactor_Cost']
```

# Reactor Heat Effect

```python
# Parameters
U                                               = 113.565
Fc                                              = 3.7935

for i in range(len(df_dict.keys())):
    # Extract data from dictioanry
    j                                           = list(df_dict.keys())[i]
    df                                          = df_dict[j]

    # Parameters for each condition
    if i <= 5:
        TLM                                         = ((T[i]-288)-(T[i]-308))/np.log((T[i]-288)/(T[i]-308))
    else:
        TLM                                         = ((T[-1]-288)-(T[-1]-308))/np.log((T[-1]-288)/(T[-1]-308))
        
    Area                                        = -df_dict[j]['REACTORQ']*10**6/(U*TLM)*10.764

    # Installation Cost
    installation_cost                           = MandS/280*101.3*Area**0.65*(2.29+Fc)
    total_capital_investment                    = 2.09*installation_cost
    annual_capital_investment                   = 1/3*total_capital_investment

    # Create a dataframe for reactor heating cost
    Reactor_Heating_Cost = pd.DataFrame({
        'Reactor_Heating_Cost'                  : {},
        'XTOL'                                  : df['XTOL'],
        'installation_cost'                     : installation_cost,
        "total_capital_investment"              : total_capital_investment,
        "annual_capital_investment"             : annual_capital_investment,
        'annual_reactor_heating_cost'           : annual_capital_investment
    })

    # Save the dataframe in the dictionary
    EP[j]['Reactor_Heating_Cost']               = Reactor_Heating_Cost

# Display 
EP['T=820K,P=6atm']['Reactor_Heating_Cost']
```

# Cooling Water Cost

```python
# Parameters
density_water =	999.13
Cp_water = 4188.60
price_coolingwater = 8.12
delta_temp = 20.00

for i in df_dict.keys():
    # Extract data from dictionary
    df = df_dict[i]

    # Calculate parameter
    volume_flowrate = -1*df['REACTORQ']*10**6/(density_water*Cp_water*delta_temp)*3600*plant_runtime
    annual_water_cost = volume_flowrate/1000*price_coolingwater

    # Create a dataframe for water cost
    Water_Cost = pd.DataFrame({
        'Water_Cost'                            : {},
        'XTOL'                                  : df['XTOL'],
        'volume_flowrate'                       : volume_flowrate,
        "annual_water_cost"                     : annual_water_cost
    })

    # Save the dataframe in the dictionary
    EP[i]['Water_Cost']               = Water_Cost

EP['T=820K,P=6atm']['Water_Cost']
```

# Furnace Heating Cost

```python
# Parameters
price_fuel                                      = 3.8
Fc                                              = 1.75

for i in df_dict.keys():
    df = df_dict[i]
    Furnace_Q                                   = df['FURNACEQ']*(3412141.6/10**6)            # in UNIT: BTU * 10^6 

    # Annual Operating Cost
    annual_operating_cost                       = Furnace_Q*price_fuel*plant_runtime

    # installation cost
    installation_cost                           = MandS/280 * 5.52*10**3 * Furnace_Q**0.85 * (1.27+Fc)
    total_capital_investment                    = 2.09*installation_cost
    annual_capital_investment                   = 1/3*total_capital_investment

    # Total annual furnace cost
    annual_furnace_cost                         = annual_operating_cost+annual_capital_investment

    # Create a dataframe for furnace cost
    Furnace_Cost = pd.DataFrame({
        'Furnace_Cost'                          : {},
        'XTOL'                                  : df['XTOL'],
        'annual_operating_cost'                 : annual_operating_cost,
        'installation_cost'                     : installation_cost,
        "total_capital_investment"              : total_capital_investment,
        "annual_capital_investment"             : annual_capital_investment,
        'annual_furnace_cost'                   : annual_furnace_cost
    })

    # Save the dataframe in the dictionary
    EP[i]['Furnace_Cost']                       = Furnace_Cost

# Display 
EP['T=820K,P=6atm']['Furnace_Cost']
```

# Pump Cost

```python
# Parameters
molar_volume = {
    'Toluene'                                   : 0.105993113,
    'Methanol'                                  : 0.040330365,
    'Benzene'                                   : 0.088961406,
    'Acetonitrile'                              : 0.052487604
}
pressure_drop                                   = 506625
price_electricity                               = 0.035

for i in df_dict.keys():
    df = df_dict[i]

    shaft_work                                  = (df['INTOL']*molar_volume['Toluene'] 
                                                   + df["INMETH"]*molar_volume['Methanol'] 
                                                   + df['INBENZ']*molar_volume['Benzene'] 
                                                   + df['INACET']*molar_volume['Acetonitrile']) * pressure_drop / (3600)

    shaft_work_hp = shaft_work/745.7

    # Annual Operating Cost 
    annual_operating_cost                       = shaft_work_hp * price_electricity * plant_runtime

    # Installation Cost
    installation_cost                           = (MandS/1415)*(-1100+2100*shaft_work_hp**0.6)
    total_capital_investment                    = 2.09*installation_cost
    annual_capital_investment                   = 1/3*total_capital_investment

    # Total annual furnace cost
    annual_pump_cost                           = annual_operating_cost+annual_capital_investment

    # Create a dataframe for furnace cost
    Pump_Cost = pd.DataFrame({
        'Pump_Cost'                             : {},
        'XTOL'                                  : df['XTOL'],
        'annual_operating_cost'                 : annual_operating_cost,
        'installation_cost'                     : installation_cost,
        "total_capital_investment"              : total_capital_investment,
        "annual_capital_investment"             : annual_capital_investment,
        'annual_pump_cost'                      : annual_pump_cost
    })

    # Save the dataframe in the dictionary
    EP[i]['Pump_Cost']                         = Pump_Cost

# Display 
EP['T=820K,P=6atm']['Pump_Cost']
```

# Economic Potential 2

## Parmeters

```python
# Parameters
molar_mass = {
    'Ethylene'                                          : 28.05,
    'Methanol'                                          : 32.04,
    'Benzene'                                           : 78.11,
    'Acetonitrile'                                      : 41.05,
    'Water'                                             : 18.02,
    'Toluene'                                           : 92.14,
    'p-xylene'                                          : 106.16,
    'm-xylene'                                          : 106.16,
    'o-xylene'                                          : 106.16
}

price_component = {
    'Toluene'                                           : 0.68, 
    'Methanol'                                          : 0.38, 
    'p-xylene'                                          : 1.31, 
    'o-xylene'                                          : 0.67,
    'm-xylene'                                          : 0.70,
    'Benzene'                                           : 0.83
}

combustionheat = {
    'p-xylene'                                          : -4313.74408, 
    'o-xylene'                                          : -4313.74408, 
    'm-xylene'                                          : -4311.84834, 
    'Benzene'                                           : -3096.68246,
    'Ethylene'                                          : -1337.44076,
    'Acetonitrile'                                      : -1190.834123
}

purge_ratio                                             = 0.1
```

## Base Case

```python
# Base Case
for i in df_dict.keys():
    # Extract data from dictionary
    df                                                  = df_dict[i]
    TOLFEED                                             = df_dict[i]['TOLFEED']
    METHFEED                                            = df_dict[i]['METHFEED']

    OUTBENZ                                             = df_dict[i]['OUTBENZ']
    OUTMETH                                             = df_dict[i]['OUTMETH']
    OUTPXYL                                             = df_dict[i]['OUTPXYL']
    OUTMXYL                                             = df_dict[i]['OUTMXYL']
    OUTOXYL                                             = df_dict[i]['OUTOXYL']
    OUTETH                                              = df_dict[i]['OUTETH']

    # Calculate economic potential for each component
    TOL                                                 = -1*TOLFEED*price_component['Toluene']*molar_mass['Toluene']
    BENZ                                                = OUTBENZ*price_component['Benzene']*molar_mass['Benzene']
    METH                                                = (purge_ratio*OUTMETH - METHFEED)*price_component['Methanol']*molar_mass['Methanol']
    PXYL                                                = OUTPXYL*price_component['p-xylene']*molar_mass['p-xylene']
    MXYL                                                = OUTMXYL*price_component['m-xylene']*molar_mass['m-xylene']
    OXYL                                                = OUTOXYL*price_component['o-xylene']*molar_mass['o-xylene']
    ETH                                                 = OUTETH*(-combustionheat['Ethylene'])*1000*price_fuel/10**6

    # Calculate economic potential
    economic_potential_2                                = (TOL+BENZ+METH+PXYL+MXYL+OXYL+ETH)*plant_runtime

    # Save the dataframe in the dictionary
    EP2 = pd.DataFrame({
        'Economic Potential 2'                          : {},
        'XTOL'                                          : df['XTOL'],
        'TOLCOST'                                       : TOL,
        'BENZCOST'                                      : BENZ,
        'METHCOST'                                      : METH,
        'PXYLCOST'                                      : PXYL,
        'MXYLCOST'                                      : MXYL,
        'OXYLCOST'                                      : OXYL,
        'ETHCOST'                                       : ETH,
        'economic_potential_2'                          : economic_potential_2
    })

    EP[i]['EP2']                                        = EP2

EP['T=820K,P=6atm']['EP2']
```

# Economic Potential 3

## Sensitivity Analysis on Temperature and Pressure

```python
for i in df_dict.keys():

    # Extract data from dictionary
    economic_potential_2                        = EP[i]['EP2']['economic_potential_2']
    annual_reactor_cost                         = EP[i]['Reactor_Cost']['annual_reactor_cost']
    annual_reactor_heating_cost                 = EP[i]['Reactor_Heating_Cost']['annual_reactor_heating_cost']
    annual_furnace_cost                         = EP[i]['Furnace_Cost']['annual_furnace_cost']
    annual_pump_cost                            = EP[i]['Pump_Cost']['annual_pump_cost']
    annual_water_cost                           = EP[i]['Water_Cost']['annual_water_cost']

    # Calculate Economic Potential 3
    economic_potential_3                        = (economic_potential_2
                                                   - annual_reactor_cost
                                                   - annual_reactor_heating_cost
                                                   - annual_furnace_cost
                                                   - annual_pump_cost
                                                   -annual_water_cost)
    
    # Save the dataframe in the dictionary
    EP3 = pd.DataFrame({
        'Economic Potential 3'                  : {},
        'XTOL'                                  : df['XTOL'],
        'economic_potential_2'                  : economic_potential_2,
        'annual_reactor_cost'                   : annual_reactor_cost,
        'annual_reactor_heating_cost'           : annual_reactor_heating_cost,
        'annual_furnace_cost'                   : annual_furnace_cost,
        'annual_pump_cost'                      : annual_pump_cost,
        'annual_water_cost'                     : annual_water_cost,
        'economic_potential_3'                  : economic_potential_3
    })

    EP[i]['EP3']                                = EP3

EP['T=820K,P=6atm']['EP3'] 
```
